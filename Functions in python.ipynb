```python
def add_sub_mul(x,y):
    c=x+y
    d=x-y
    e=x*y
    return c,d,e
    
result,result1,result2 =add_sub_mul (3,4)  

print(result)
print(type(result))

print(result1)
print(type(result1))

print(result2)
print(type(result2))
```
